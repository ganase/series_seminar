```python
import numpy as np
import pandas as pd
from scipy import stats

from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
sns.set()

import statsmodels.api as sm
```

```python
#ホワイトノイズ
mean = 0
sigma = 1 
T = 100

epsilons = np.random.normal(mean, sigma, size=T)

plt.figure(figsize=(12,8))
plt.plot(epsilons)
```

```python
#ランダムウォーク1
mu = 0
sigma = 1
T = 100

cepsilons = np.random.choice([1, -1], size=T)

crandom = np.zeros(T)
crandom[0] = mu + cepsilons[0]

for t in range(1, T):
    crandom[t] = crandom[t-1] + cepsilons[t]

plt.figure(figsize=(12,8))
plt.plot(crandom)
```

```python
#ランダムウォーク2
mu = 0
sigma = 1
T = 100

epsilons = np.random.normal(0, sigma, size=T)

random = np.zeros(T)
random[0] = mu + epsilons[0]

for t in range(1, T):
    random[t] = random[t-1] + epsilons[t]

plt.figure(figsize=(12,8))
plt.plot(random)
```

```python
#ランダムウォークの差分
diffrandom = np.diff(random)

plt.figure(figsize=(12,8))
plt.plot(diffrandom)
```

```python
#ランダムウォーク3
mu = 0
sigma = 1
T = 100
c = 0.5

epsilons = np.random.normal(0, sigma, size=T)

drandom = np.zeros(T)
drandom[0] = mu + epsilons[0]

for t in range(1, T):
    drandom[t] = c + drandom[t-1] + epsilons[t]

plt.figure(figsize=(12,8))
plt.plot(drandom)
```
